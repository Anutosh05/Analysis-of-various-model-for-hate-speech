### Reasons for Recall 0 for Many Classes

The **Recall** value of 0 for certain classes indicates that the model is failing to identify instances of those classes during evaluation. Possible reasons for this include:

#### 1. **Imbalanced Dataset**
   - If the dataset contains very few samples for a particular class, the model may not be able to learn its features effectively, resulting in a **zero recall** for that class.

#### 2. **Model Overfitting**
   - If the model overfits on the majority class, it may become biased towards predicting the majority class, leading to poor performance on the minority class, and hence zero e clas3es.

#### 4. **Thresholding Issue**
   - The model may predict low probabilities for the affected class, resulting in misclassifications. This is common in **multi-class classification** where probabilities for different classes are close tsh zero recall can be improved.


```python
model_names = df["Model Name"]
recall_class_0 = df["Recall_Class_0"]

# Create a DataFrame for plotting
plot_data = pd.DataFrame({
    "Model Name": model_names,
    "Recall (Class 0)": recall_class_0
})

# Set Seaborn style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x="Model Name", y="Recall (Class 0)", data=plot_data, palette="Blues_d", edgecolor="black")

# Add labels and title
plt.xlabel("Model Name", fontsize=12)
plt.ylabel("Recall (Class 0)", fontsize=12)
plt.title("Recall (Class 0) vs Model Name", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
class_columns = [col for col in df.columns if col.startswith("Recall_Class_")]
df["Mean Recall"] = df[class_columns].mean(axis=1)

# Extract relevant data
model_names = df["Model Name"]
mean_recall = df["Mean Recall"]

# Create a DataFrame for plotting
plot_data = pd.DataFrame({
    "Model Name": model_names,
    "Mean Recall": mean_recall
})

# Set Seaborn style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x="Model Name", y="Mean Recall", data=plot_data, palette="Greens_d", edgecolor="black")

# Add labels and title
plt.xlabel("Model Name", fontsize=12)
plt.ylabel("Mean Recall", fontsize=12)
plt.title("Mean Recall vs Model Name", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
