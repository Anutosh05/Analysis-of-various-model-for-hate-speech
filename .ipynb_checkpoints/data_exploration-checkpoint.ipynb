# Analysis of Most words used in the text

```python
all_text = " ".join(df['Content'])

# Create WordCloud object
wordcloud = WordCloud(width=1000, height=600, background_color='white').generate(all_text)

# Display the WordCloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.title("Most used word in the Dataset", fontsize=16)
plt.show()
```

```python
df['content_length'] = df['Content'].apply(len)

# Plot histogram with bins of size 10 up to 300
plt.hist(df['content_length'], bins=range(0, 310, 5), edgecolor='black')

# Set the title and labels
plt.title("Content Length Distribution")
plt.xlabel("Length")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

```python
df['word_count'] = df['Content'].apply(lambda x: len(x.split()))
print('Information about Length of text')
print(df['word_count'].describe())
```

```python
positive_text = " ".join(df[df['Label'] == 1]['Content'])
negative_text = " ".join(df[df['Label'] == 0]['Content'])

# Generate word clouds
positive_wordcloud = WordCloud(background_color='white').generate(positive_text)
negative_wordcloud = WordCloud(background_color='white').generate(negative_text)

# Set up the matplotlib plot with 2 subplots
plt.figure(figsize=(12, 6))

# Positive WordCloud
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.imshow(positive_wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis labels
plt.title("Positive Content WordCloud")

# Negative WordCloud
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.imshow(negative_wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis labels
plt.title("Negative Content WordCloud")

# Show the plot
plt.tight_layout()
```

```python
df['word_count'] = df['Content'].apply(lambda x: len(str(x).split()))
df['char_count'] = df['Content'].apply(lambda x: len(str(x)))
```

```python
correlations = df[['word_count', 'char_count', 'Label']].corr()
correlations['Label']
```

```python
print(df.groupby('Label')[['word_count', 'char_count']].mean())
sns.boxplot(x='Label', y='word_count', data=df)
plt.title("Word Count by Label")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features=100, stop_words='english')  # Ignore common stop words
word_matrix = vectorizer.fit_transform(df['Content'])
top_words = vectorizer.get_feature_names_out()

# Create a DataFrame for word occurrences
word_features = pd.DataFrame(word_matrix.toarray(), columns=top_words)
word_features['Label'] = df['Label']

# Step 2: Calculate word frequency for Label=1 and Label=0
word_freq_label1 = word_features[word_features['Label'] == 1].drop('Label', axis=1).sum() / len(word_features[word_features['Label'] == 1])
word_freq_label0 = word_features[word_features['Label'] == 0].drop('Label', axis=1).sum() / len(word_features[word_features['Label'] == 0])

# Combine results into a DataFrame for comparison
word_effect = pd.DataFrame({
    'Word': top_words,
    'Frequency_Label1': word_freq_label1,
    'Frequency_Label0': word_freq_label0
}).sort_values(by='Frequency_Label1', ascending=False)

# Step 3: Visualize the effect of words on Label=1
plt.figure(figsize=(10, 8))
plt.barh(word_effect['Word'].head(20), word_effect['Frequency_Label1'].head(20), color='green', alpha=0.7)
plt.title("Top 20 Words Associated with Label=1")
plt.xlabel("Frequency in Label=1")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# Step 4: Visualize the effect of words on Label=0
plt.figure(figsize=(10, 8))
plt.barh(word_effect['Word'].head(20), word_effect['Frequency_Label0'].head(20), color='red', alpha=0.7)
plt.title("Top 20 Words Associated with Label=0")
plt.xlabel("Frequency in Label=0")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
